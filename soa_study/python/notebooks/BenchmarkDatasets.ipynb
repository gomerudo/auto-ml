## Summary of benchmark datasets

```
# For matrices/arrays and data frames
import numpy as np
import pandas as pd

# For plotting
import matplotlib as mpl
import matplotlib.pyplot as plt

# For openml interaction
import openml as oml
from openml import tasks, runs, datasets

################################################################################
##### Function to get the information of the dataset in the form of a dict #####
################################################################################
def getDatasetProperties(dataset) :
  X, y = dataset.get_data(target = dataset.default_target_attribute)
  
  m = X.shape[1] # Number of features
  n = X.shape[0] # Number of observations
  
  classes = len(np.unique(y)) # Number of classes
  
  return { "n" : n, "nFeatures" : m, "classes" : classes }

################################################################################
#### Function to plot the information of one dataset in one axis of plt obj ####
################################################################################
def plotDataset(datasetId, axis) :
  # Obtain the dataset from OpenML
  dataset = oml.datasets.get_dataset(datasetId)\
  # Get the properties we want
  datasetProps = getDatasetProperties(dataset)
  # Obtain the y column only
  _, y = dataset.get_data(target = dataset.default_target_attribute)
  # Convert to dataframe with column 'target'
  y = pd.DataFrame(y, columns = [ "target" ])
  
  # Plot the histogram of the classes in the 'target' with a custom title
  customTitle = """Dataset : {} ({})
nObs = {} ; nFeatures = {} ; nClasses = {}""".format(dataset.name, datasetId, 
                                                     datasetProps['n'], 
                                                     datasetProps['nFeatures'], 
                                                     datasetProps['classes'])
  
  y["target"].value_counts(sort = False).plot(colormap = "tab20c",
                                                    ax = axis,
                                                    kind = 'bar', 
                                                    title = customTitle,
                                                    figsize = (22, 30))
  

################################################################################
##################################### MAIN #####################################
################################################################################
openMLIds = [ 38, 46, 179, 184, 554, 772, 917, 1049, 1111, 1120, 1128, 293, 389,
             1597]

# Define the grid to plot into
fig, axes = plt.subplots(nrows = 5, ncols = 3)
# Setting a super title to the main plot
plt.suptitle('Summary of benchmark datasets')

# For each axis in the grid, plot the corresponding dataset from the list of IDs
for i, axis in enumerate(axes.flatten()) :
  if i < len(openMLIds) : plotDataset(openMLIds[i], axis)
```
