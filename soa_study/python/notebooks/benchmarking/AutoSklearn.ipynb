## Running the classifiers on the test datasets

```
import openml as oml
from openml import tasks, runs, datasets
from sklearn.model_selection import train_test_split
import time
import autosklearn.classification
from sklearn.metrics import roc_auc_score
from autosklearn.metrics import roc_auc
from autosklearn.metrics import accuracy


################################################################################
########### Function to run a model and return the results as a list ###########
################################################################################
def runModel(dataset, metric = accuracy) :
  
  # Get the features and the target
  X, y = dataset.get_data(target = dataset.default_target_attribute)
  # Obtain the train and validation sets with a 3/4 split
  X_train, X_val, y_train, y_val = train_test_split(X, y, 
                                                  train_size = 0.75, 
                                                  test_size = 0.25) 
  
  print("""
================================================================================
RUNNING TPOT CLASSIFIER FOR DATASET {}
================================================================================
  """.format(dataset.dataset_id))
  cls = autosklearn.classification.AutoSklearnClassifier()
  
  start_time = time.time()
  cls.fit(X_train, y_train, metric = metric, dataset_name = str(dataset.dataset_id))
  end_time = time.time()

  
  return [ dataset.dataset_id, end_time - start_time, 
          roc_auc_score(y_val, cls.predict(X_val)), cls.sprint_statistics(), cls.get_models_with_weights() ]
```

### Part 1

```
benchmarks_metrics = { 38 : roc_auc, 46 : accuracy } 
```

```
################################################################################
##################################### MAIN #####################################
################################################################################
results = []

for datasetId, metric in benchmarks_metrics.items():
  dataset = oml.datasets.get_dataset(datasetId) 
  results.append(runModel(dataset, metric))
```

```
import pandas as pd
pd.set_option('display.width', 1024)
pd.set_option('display.max_colwidth', 1000)

pd.DataFrame(results, columns = ["Dataset id", "Time", "Validation score", "Statistics", "Pipeline"])
```
