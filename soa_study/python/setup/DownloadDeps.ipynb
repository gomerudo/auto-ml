## Option 1: pip download

### 1. Download whl files with pip download

```
!rm -rf ./pypi_home ./pip_build
```

```
!cat pip.txt | xargs -n 1 pip3.4 download \
  --only-binary=:all: \
  --platform Linux-4.4.121-92.92-default-x86_64-with-SuSE-12-x86_64 \
  --python-version 346 \
  --implementation cp \
  --dest ./pypi_home \
  --build ./pip_build
```

### 2. Download dependencies in tar.gz form

```
!cat targz.txt | xargs -n 1 pip3.4 download \
  --dest ./pypi_home \
  --build ./pip_build 
```

### Creating subdirectories to build a proper local pypi

```
!find pypi_home/ -maxdepth 1 -type f | xargs -n 1 basename | cut -d '-' -f 1 | xargs -I {} mkdir -p pypi_home/{}
```

```
!find pypi_home/ -maxdepth 1 -type d | xargs -n 1 basename | while read -r line ; do echo ${line} ; find pypi_home -iname ${line}* -type f | xargs -I {} mv {} pypi_home/${line}/.; done
```

```
!ls pypi_home
```

### Zipping the pypi_home

```
! rm -rf pypi_home.zip; zip -r pypi_home.zip pypi_home
```

```
!rm -rf aux_pypi; mkdir -p aux_pypi; split -b 20M pypi_home.zip "aux_pypi/pypi_home.zip.part"
```

```
!ls aux_pypi
```
